```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("processed.cleveland.data", header = None)
df
```

```python
df.columns = ["Age",
"Gender",
"cp",
"restbp",
"chol",
"fbs",
"restecg",
"thalach",
"exang",
"oldpeak",
"slope",
"ca",
"thal",
"hd"]
```

```python
df.head()
```

```python
df.info()
```

```python
df["ca"].unique()
```

```python
df.thal.unique()
```

```python
df[(df["ca"] == "?") | (df["thal"] == "?")]
```

```python
df_no_missing = df[(df["ca"] != "?") & (df["thal"] != "?")]
df_no_missing
```

```python
x = df_no_missing.drop("hd", axis = 1)
y = df_no_missing["hd"]
```

```python
df["cp"].unique()
```

```python
x_encoded = pd.get_dummies(x, columns = ["cp",
"restecg",
"slope",
"thal"], dtype = int)
```

```python
x_encoded.shape
```

```python
x_encoded
```

```python
y.unique()
```

```python
y[y>0] = 1
y.unique()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, random_state=42)
```

```python
rf = RandomForestClassifier(random_state = 42)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
y_pred_train = rf.predict(x_train)
print(confusion_matrix(y_pred_train, y_train))
print(classification_report(y_pred_train, y_train))
```

```python
params = {
"max_depth": [2,3,5,7],
"min_samples_leaf" : [5,10,15,20],
"n_estimators": [10,15,20]
}
```

```python
gridsearch = GridSearchCV(estimator = rf,
param_grid = params,
cv = 4,
scoring = "accuracy")
gridsearch.fit(x_train, y_train)
```

```python
gridsearch.best_estimator_
```

```python
params = {
"max_depth": [2,3],
"min_samples_leaf" : [5,10,15],
"n_estimators": [12,15,17]
}
gridsearch = GridSearchCV(estimator = rf,
param_grid = params,
cv = 4,
scoring = "accuracy")
gridsearch.fit(x_train, y_train)
gridsearch.best_estimator_
```

```python
params = {
"max_depth": [2,3],
"min_samples_leaf" : range(1,20),
"n_estimators": range(15,17)
}
gridsearch = GridSearchCV(estimator = rf,
param_grid = params,
cv = 4,
scoring = "accuracy")
gridsearch.fit(x_train, y_train)
gridsearch.best_estimator_
```

```python
y_pred = gridsearch.predict(x_test)
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
y_pred_train = gridsearch.predict(x_train)
print(confusion_matrix(y_pred_train, y_train))
print(classification_report(y_pred_train, y_train))
```

```python
rf_best=gridsearch.best_estimator_
rf_best
```

```python
rf_best.feature_importances_
```

```python
imp_df = pd.DataFrame({
"Varname" : x_train.columns,
"Importance" : rf_best.feature_importances_
})
imp_df
```

```python
imp_df_sorted = imp_df.sort_values(by = "Importance", ascending = False)
imp_df_sorted.reset_index()
```
